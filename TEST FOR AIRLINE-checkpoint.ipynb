```python
from factor_analyzer import FactorAnalyzer
```

```python
df = pd.read_csv('train.csv')
```

```python
gendermapping = {'Male': 0, 'Female': 1}
df['Gendernumber'] = df['Gender'].map(gendermapping)

classmapping = {'Eco': 0, 'Eco Plus': 1, 'Business': 2}
df['Classnumber'] = df['Class'].map(classmapping)


satisfactionmapping = {'neutral or dissatisfied': 0, 'satisfied': 1}
df['Satisfactionnumber'] = df['satisfaction'].map(satisfactionmapping)

traveltypemapping = {'Personal Travel': 0, 'Business travel': 1}
df['traveltypenumber'] = df['Type of Travel'].map(traveltypemapping)

customertypemapping = {'disloyal Customer': 0, 'Loyal Customer': 1}
df['Customertypenumber'] = df['Customer Type'].map(customertypemapping)
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0', 'id', 'Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction'], axis = 1, inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
fa = FactorAnalyzer(n_factors=5, rotation ='varimax')
```

```python
df = df.dropna()
```

```python
fa.fit(df)
```

```python
#get the loadings
loadings = fa.loadings_
```

```python
#get the Eigenvector and eigenvalues
ev, v = fa.get_eigenvalues()
```

```python
xvals = range (1,df.shape[1]+1)
```

```python
plt.scatter(xvals,ev)
plt.plot(xvals, ev)
plt.title ('Scree Plot')
plt.xlabel ('Factor')
plt.ylabel ('Eigenvalue')
plt.grid()
plt.show()
```
